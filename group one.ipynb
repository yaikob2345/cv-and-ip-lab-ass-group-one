```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Read the image
image = mpimg.imread('cover page.png')  # Replace 'cover page.png' with your image path

# Create a figure and set the window size
plt.figure(figsize=(30, 10))  # Set the size of the window (width, height)

# Display the image
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```

# Load the color image (Lena image)


```python
lena_color = cv2.imread(cv2.samples.findFile('lena image.jpeg'))
lena_color = cv2.cvtColor(lena_color, cv2.COLOR_BGR2RGB)  # Convert to RGB for display

```

# Convert to grayscale


```python
lena_gray = cv2.cvtColor(lena_color, cv2.COLOR_RGB2GRAY)

```

```python
## Convert grayscale to binary and
## Display all three images in a single window
```

```python

_, lena_binary = cv2.threshold(lena_gray, 128, 255, cv2.THRESH_BINARY)


plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(lena_color)
plt.title('Color Image (Lena)')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(lena_gray, cmap='gray')
plt.title('Grayscale Image (lena)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(lena_binary, cmap='gray')
plt.title('Binary Image lena')
plt.axis('off')

plt.tight_layout()
plt.show()
```
